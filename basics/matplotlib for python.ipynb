```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pyplot
```

```python
x = np.arange(0,11)
x
```

```python
y = x**2
pyplot.style.use('bmh')
test = pyplot.plot(x,y,"gx")
pyplot.xlim(0,5)
pyplot.ylim(0,10)
pyplot.xlabel("this is x label")
pyplot.ylabel("this is y label")
pyplot.title("title for the graph")

pyplot.show(test)
```

## Second Example - graph styles

```python
matrix = np.arange(0,1000).reshape(10,100)
matrix
```

```python
pyplot.show(pyplot.contourf(matrix,cmap='ocean_r'),pyplot.colorbar())
```

```python
gmMat = np.random.randint(0,100,(10,10))
gmMat
```

```python
gmPlot = pyplot.imshow(gmMat,cmap='winter_r')

def gmShow(gmvar,gmcolorBar=''):
    pyplot.show(gmvar)

gmShow(gmPlot)
```
